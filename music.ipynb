```python
import matplotlib.pyplot as plt
import seaborn as sns

features = ['energy',  'danceability', 'tempo','valence','followers','popularity_artists']

plt.figure(figsize=(16, 12))
for i, col in enumerate(features, 1):#to get both index and column name
    plt.subplot(4, 2, i)
    sns.histplot(df[col], kde=True, bins=30, color='mediumslateblue')
    plt.title(f'Distribution of {col}')
    
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import  MinMaxScaler

# Select numeric features for clustering


# Extract feature matrix
X = df[features]

# Choose scaler
scaler =MinMaxScaler()

# Fit and transform
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

sse = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, sse, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("SSE (Inertia)")
plt.title("🦾 Elbow Method for Optimal k")
plt.grid(True)
plt.show()
```

```python
# Choose optimal k based on elbow (e.g., k=5)
optimal_k = 4
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans_final.fit_predict(X_scaled)

# Add cluster labels to original DataFrame
df['cluster'] = cluster_labels
```

```python
print(df['cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 1: Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 2: Add PCA components to DataFrame (optional)
df['PC1'] = X_pca[:, 0]
df['PC2'] = X_pca[:, 1]

# Step 3: Visualize with cluster labels
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='tab10', alpha=0.6, edgecolor='k')
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)")
plt.title("🎵 PCA Projection of Music Features")
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Step 1: Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Step 2: Visualize
plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=cluster_labels, cmap='tab10', alpha=0.6, edgecolor='k')
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("🎵 t-SNE Projection of Music Features")
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

score = silhouette_score(X_scaled, kmeans.labels_)
print(f"Silhouette Score: {score:.3f}")


db_index = davies_bouldin_score(X_scaled, kmeans.labels_)
print(f"Davies-Bouldin Index: {db_index:.3f}")
print(f"Inertia (SSE): {kmeans.inertia_:.2f}")

```

```python
cluster_profiles = df.groupby('cluster').mean(numeric_only=True)
```

```python
def sample(rows):
    if rows['danceability'] > 0.7 and rows['energy'] > 0.7:
        return 'Energetic Dance'    
    if rows['energy'] > 0.7 and rows['valence'] > 0.6:
        return 'Happy Energetic'
    elif rows['valence'] > 0.7 and rows['tempo'] > 100:
        return "😊 Feel-Good Pop"
    elif rows['valence'] < 0.4 and rows['danceability'] < 0.5:
        return "😔 Melancholy Ballads"
    elif rows['instrumentalness'] > 0.6:
        return "🎼 Instrumental Focus"
    else:
        return "🎶 Mixed Mood"
cluster_profiles['description'] = cluster_profiles.apply(sample, axis=1)
```
